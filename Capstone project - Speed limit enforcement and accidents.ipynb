# Traffic Accidents vs Nightlife and the relocation of Traffic Enforcement Cameras according to accident data in Medellín

## Introduction

In this project, we will analyze some of the patterns of traffic accidents in Medellín, Colombia. The project will be divided in two parts:

In the first part, we will check if there is any relationship between traffic accidents and  nightlife. We will see if there exist evidence that supports the hypotesis that there are more serious accidents in the weekends, specially in late night/ early morning hours. Using the foursquare API, we will also check close venues to incidents that involve deaths in those hours, in order to see if there are bars or clubs among those venues, which may suggest drunk drivers causing these deaths.

The second part of he project consists in a comparison between traffic accidents data and the geolocation of speed cameras in the city, to see if there are high risk zones that are not properly covered and can be intervened by the authorities. We will apply the K-means method to the traffic accidents dataset, to cluster the incidents and find their centroids, which in this case we will assume represent high risk areas. We will compare these locations to the current traffic enforcement camera locations, in order to suggest a new location for the cameras towards the goal of making traffic safer.

Next, we pick a search radius for venues and create a function to retrieve a dataframe with venues nearby each point:

```python
LIMIT = 100 
radius = 500 
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url3 = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url3).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

We call the function with the points in the fatal accidents dataset:

```python

sl_close_venues = getNearbyVenues(names=df_wknd['OBJECTID'],
                                   latitudes=df_wknd['Latitude'],
                                   longitudes=df_wknd['Longitude']
                                  )
```

Each ID will get nearby locations. We apply one-hot encoding for each venue type and define a function to group 10 most common venue category close to each point:

```python
# one hot encoding
med_onehot = pd.get_dummies(sl_close_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
med_onehot['Neighborhood'] = sl_close_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [med_onehot.columns[-1]] + list(med_onehot.columns[:-1])
med_onehot = med_onehot[fixed_columns]

med_grouped = med_onehot.groupby('Neighborhood').mean().reset_index()
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = med_grouped['Neighborhood']

for ind in np.arange(med_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(med_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

As we can see, out of the 24 fatal accident locations, there are only three where the closest venue are bars, and there is only one point where the closest venue is a Nightclub.

This doesn't bring any strong evidence that these incidents might be caused by drunk drivers. Moreover, even if the driver was in fact drunk, the accident location might not be close to the point where the person got drunk, or the driver could have drunk in a different place, not a bar or nightclub. More data needs to be retrieved in order to have a deeper understanding of the possible causes of each accident. 

# Part 2: Traffic accidents and speed cameras

In the second part of this poject, we will use a second dataset, which contains the coordinates of all the traffic enforcement cameras (speed cameras) in the city. We will apply a K-means model of n cluster of accidents, where n is the number of working speed cameras in the city. The K-means centroids will give us the potential new locations of the cameras, according to accident concentration.

First, we import our second dataset, and check the number of cameras:

```python
url="https://www.datos.gov.co/resource/js6n-qi6v.csv"
df=pd.read_csv(url)
df.columns=["Id", "Address", "Max Speed", "Longitude", "Latitude", "Coordinates"]
df=df.drop(columns=['Id'])
df.shape
```

There's 77 cameras currently working. For the next step, we'll create a dataframe containing all the accidents that involve injuries or deaths, and check its size:

```python
df3=df2[~df2.GRAVITY.str.contains("SOLO DAÑOS")].reset_index().drop(columns=['index'])
df3=df3[["Longitude", "Latitude"]]
df3.shape
```

There were 23217 accidents with said characteristics. Now we can normalize our dataset:

```python
from sklearn.preprocessing import StandardScaler
X = df3.values[:,1:]
X = np.nan_to_num(X)
Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet
```

### Modeling

Let's run our model and cluster the accidents into 77 clusters, one for each existing camera:

```python
clusterNum = 77
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

Finally, we'll create the centroids dataframe, that has the coordinates of the new suggested locations for each speed camera:

```python
df3["Cluster"] = labels
```

```python
centroids= df3.groupby('Cluster').mean()
centroids['Cluster'] = centroids.index
centroids.head()
```
